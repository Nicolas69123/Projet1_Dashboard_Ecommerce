```python
import duckdb
import pandas as pd

# Connexion à la base
con = duckdb.connect("ecommerce.duckdb")

# Charger le CSV nettoyé
df = pd.read_csv("donnees_nettoyees.csv", parse_dates=["InvoiceDate"])

# Créer table transactions
con.execute("""
CREATE TABLE IF NOT EXISTS fact_sales (
    InvoiceNo VARCHAR,
    StockCode VARCHAR,
    Description VARCHAR,
    Quantity INTEGER,
    InvoiceDate TIMESTAMP,
    UnitPrice DOUBLE,
    CustomerID DOUBLE,
    Country VARCHAR,
    TransactionType VARCHAR,
    IsIdentifiedCustomer BOOLEAN,
    TotalAmount DOUBLE
)
""")

# Insertion données
con.register("df", df)
con.execute("INSERT INTO fact_sales SELECT * FROM df")

print("✅ Base et table créées avec succès")
```
